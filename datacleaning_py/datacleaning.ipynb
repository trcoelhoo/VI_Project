```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel('PORDATA/Portugal/recintos_ecras_sessoes_espectadores_portugal.xlsx',skiprows=6)
```

```python
data 
```

```python
dados_2011= data.loc[data['Anos'].values == [2011]]
dados_2012= data.loc[data['Anos'].values == [2012]]
dados_2013= data.loc[data['Anos'].values == [2013]]
dados_2014= data.loc[data['Anos'].values == [2014]]
dados_2015= data.loc[data['Anos'].values == [2015]]
dados_2016= data.loc[data['Anos'].values == [2016]]
dados_2017= data.loc[data['Anos'].values == [2017]]
dados_2018= data.loc[data['Anos'].values == [2018]]
dados_2019= data.loc[data['Anos'].values == [2019]]
dados_2020= data.loc[data['Anos'].values == [2020]]
dados_2021= data.loc[data['Anos'].values == [2021]]

data=pd.concat([dados_2011, dados_2012,dados_2013,dados_2014,dados_2015,dados_2016,dados_2017,dados_2018,dados_2019,dados_2020,dados_2021], axis=0)
data
```

```python
df1 = pd.DataFrame(data['Anos'])
df2 = pd.DataFrame(data['Espectadores'])
new_data=pd.concat([df1, df2], axis=1)
new_data
```

```python
new_data.to_csv('espectadores_portugal.csv')
```
